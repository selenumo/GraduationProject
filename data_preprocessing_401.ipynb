To manipulate the empty data following rules will imply:
If first and last data of the row is empty then fill with '0'.
If interim datas are empty then 
  If the previous and next data is not empty current data is filled with the average of previous and next data. Otherwise current data is filled with '0.'
To manipulate the erroneous data following rules will imply:
The threshold to determine the data is erroneous or not is avg of row plus two times standart deviation of row. 
If current data is bigger than the threshold then the current data is filled with avg(row)+2*st. 
The output will be saved in the output_dir with the data with above rules applied. As the new row processed then it will be appended in the file.

```python
for index, row in df[0:27580].iterrows():
  for i in range(len(row)):
    l = len(row)-1
    if (row.isnull()[0]):  
      row[0] = 0
    if (row.isnull()[l]):
      row[l] = 0
    if (i>0 and i<l and row.isnull()[i]):
      if (row.isnull()[i-1]== False and row.isnull()[i+1]== False):
        row[i] = (row[i-1]+row[i+1])/2
      else:
        row[i] = 0
  avg = np.average(row)
  st = np.std(row)
  for i in range(len(row)):
    thr = avg+2*st
    if(row[i]>thr):
      row[i] = thr
  output_dir = '/content/drive/MyDrive/ele401/data_401/'+'output.txt'
  with open(output_dir, "a") as f:
    f.write(" ".join([("%.3f" %a ) for a in row]) + '\n')
```

```python
df = pd.read_csv('/content/drive/MyDrive/ele401/data_401/data_final.csv')
df.drop(['Unnamed: 0'],axis =1,inplace=True)
Y = df['OutPut']
df.drop(['OutPut'],axis=1,inplace=True)
```

```python
mt = np.array(df)
mt.shape
```

```python
for i in range(len(mt)):
  mx = mt[i].max()
  mn = mt[i].min()
  if( (mx-mn)==0):
    continue
  for j in range(len(mt[i])):
    mt[i][j] = (mt[i][j]-mn)/(mx-mn)
```
